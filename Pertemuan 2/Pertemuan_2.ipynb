```python
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

# create node
n1 = Node("Januari")
n2 = Node("Februari")

class LinkedList:
  def __init__(self):
    self.headval = None

# Print the linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

# function to add node in the end
  def AddInEnd(self,newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      return
    last = self.headval
    while(last.nextval):
      last = last.nextval
    last.nextval=NewNode

Li = LinkedList()
Li.headval = n1

# Link first node to second node
Li.headval.nextval = n2

# add node in the end
Li.AddInEnd("The Last")
Li.listprint()
```

Menghapus node pada linked list

```python
# HAPUS NODE DENGAN KATA KUNCI

class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

  def AddBegining(self, newdata):
    NewNode = Node(newdata)
    NewNode.nextval = self.headval
    self.headval = NewNode

  def RemoveNode(self, Removekey):
    HeadVal = self.headval

    if HeadVal is not None:
      if HeadVal.dataval == Removekey:
        self.headval = HeadVal.nextval
        HeadVal = None
        return

    while HeadVal is not None:
      if HeadVal.dataval == Removekey:
        break
      prev = HeadVal
      HeadVal = HeadVal.nextval

    if HeadVal is None:
      return

    prev.nextval = HeadVal.nextval
    HeadVal = None


# ================================
# Program Utama (di luar class)
# ================================

Li = LinkedList()

Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("Februari")

print("Sebelum dilakukan remove")
Li.listprint()

print("================================")

Li.RemoveNode("Maret")

print("Setelah dilakukan remove dengan kunci 'Maret'")
Li.listprint()
```

```python
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

# create node
n1 = Node("Januari")
n2 = Node("Februari")

class LinkedList:
  def __init__(self):
    self.headval = None

# Print the linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

# function to add node in the end
  def AddInEnd(self,newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      return
    last = self.headval
    while(last.nextval):
      last = last.nextval
    last.nextval=NewNode

# Function to remove first node
  def RemoveEnd(self):
    last = self.headval
    while (last is not None):
      if last.nextval == None:
        break
      prev = last
      last = last.nextval
    if (last == None):
      return
    prev.nextval = last.nextval
    last = None

# create linked list named Li
Li = LinkedList()

# add node
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")
print("Sebelum dilakukan remove")
Li.listprint()
print("==============================================")

# remove first node
Li.RemoveEnd()
print("Setelah dilakukan remove node pertama")
Li.listprint()
```

```python
# DOUBLE LINKED LIST

class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None
    self.prevval = None


class LinkedList:
  def __init__(self):
    self.headval = None

# Print the linked list (maju)
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print(printval.dataval)
      printval = printval.nextval

# Add node di akhir
  def AddEnd(self, newdata):
    NewNode = Node(newdata)

    if self.headval is None:
      self.headval = NewNode
      return

    temp = self.headval
    while temp.nextval:
      temp = temp.nextval

    temp.nextval = NewNode
    NewNode.prevval = temp

# Insert di tengah (setelah kata kunci)
  def InsertMiddle(self, key, newdata):
    temp = self.headval

    while temp is not None:
      if temp.dataval == key:
        break
      temp = temp.nextval

    if temp is None:
      print("Data tidak ditemukan")
      return

    NewNode = Node(newdata)
    NewNode.nextval = temp.nextval
    NewNode.prevval = temp

    if temp.nextval is not None:
      temp.nextval.prevval = NewNode

    temp.nextval = NewNode

# Remove berdasarkan kata kunci
  def RemoveNode(self, key):
    temp = self.headval

    while temp is not None:
      if temp.dataval == key:
        break
      temp = temp.nextval

    if temp is None:
      print("Data tidak ditemukan")
      return

    # Jika node pertama
    if temp.prevval is None:
      self.headval = temp.nextval
      if self.headval is not None:
        self.headval.prevval = None
    else:
      temp.prevval.nextval = temp.nextval
      if temp.nextval is not None:
        temp.nextval.prevval = temp.prevval

Li = LinkedList()

# Membuat 3 node
Li.AddEnd("ayam")
Li.AddEnd("sapi")
Li.AddEnd("kucing")

print("Data awal:")
Li.listprint()
print("================================")

# Penyisipan tengah (setelah sapi)
Li.InsertMiddle("sapi", "kambing")
print("Setelah penyisipan 'kambing' setelah 'sapi':")
Li.listprint()
print("================================")

# Hapus berdasarkan kata kunci
Li.RemoveNode("sapi")
print("Setelah menghapus 'sapi':")
Li.listprint()
```

```python
# CIRCULAR LINKED LIST

class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None


class CircularLinkedList:
  def __init__(self):
    self.headval = None

# Print circular linked list
  def listprint(self):
    if self.headval is None:
      return

    printval = self.headval
    while True:
      print(printval.dataval)
      printval = printval.nextval
      if printval == self.headval:
        break

# Add node di akhir
  def AddEnd(self, newdata):
    NewNode = Node(newdata)

    # Jika list kosong
    if self.headval is None:
      self.headval = NewNode
      NewNode.nextval = self.headval
      return

    temp = self.headval
    while temp.nextval != self.headval:
      temp = temp.nextval

    temp.nextval = NewNode
    NewNode.nextval = self.headval

# Remove node di akhir
  def RemoveEnd(self):
    if self.headval is None:
      return

    # Jika hanya 1 node
    if self.headval.nextval == self.headval:
      self.headval = None
      return

    temp = self.headval
    while temp.nextval.nextval != self.headval:
      temp = temp.nextval

    temp.nextval = self.headval

Li = CircularLinkedList()

# Membuat 3 node
Li.AddEnd("merkurius")
Li.AddEnd("venus")
Li.AddEnd("bumi")

print("Sebelum dilakukan penambahan dan penghapusan")
Li.listprint()
print("==============================================")

# Penyisipan di akhir
Li.AddEnd("mars")
print("Setelah dilakukan penambahan di akhir")
Li.listprint()
print("==============================================")

# Hapus node akhir
Li.RemoveEnd()
print("Setelah dilakukan penghapusan node terakhir")
Li.listprint()
```
